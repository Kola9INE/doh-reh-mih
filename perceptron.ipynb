```python
import pandas as pd
df = pd.read_csv('low.txt', delimiter = '\t')
df.columns
```

```python
df = df[['SLOPE', 'INTERCEPT', 'LABEL']]
df.head()
```

```python
slopes = df['slope'.upper()]
intercepts = df['intercept'.upper()]
label = df['label'.upper()]

x = 0
data = []
while x < len(slopes):
    data.append((slopes[x], intercepts[x], label[x]))
    x+=1
```

```python
"""
CONCERNS:
1. SHOULD WE INCREASE OR DECREASE LEARNING RATE?
2. WHEN SHOULD WE INCREASE OR DECREASE LEARNING RATE?
3. SHOULD WE USE THE AMOUNT OF EPOCHS THAT HAVE RUN TO DETERMINE WHEN TO INCREASE OR DECREASE LEARNING RATE?
4. SH0ULD WE USE ERROR PERCENT TO DETERMINE WHEN TO INCREASE OR DECREASE LEARNING RATE?
"""

w = [0.1,0.1,0.1]
lr = 1.0
print(f'Inital weight: {w}\n')
error = [0]
epoch = 0
while error:
    error = []
    if epoch == 10000:
        break

    if (epoch > 1000) and (epoch % 1000 == 0):
        lr-=0.1
        
    epoch+=1
    print('Epoch:',epoch)
    # print(f'{'Label':10}','Y_pred')
    for record in data:
        y_pred = w[0]+(w[1]*record[0])+(w[2]*record[1])
        if y_pred > 0:
            y_pred = 1
        else:
            y_pred = 0
            
        nwt = w[0]+(record[2]-y_pred)*lr
        nwt1 = w[1]+(record[2] - y_pred) * record[0] *lr
        nwt2 = w[2]+(record[2]-y_pred)* record[1] *lr
        w[0], w[1], w[2] = nwt, nwt1, nwt2
        #print(f'{str(record[2]):10}', f'{str(y_pred)}')
        # print(w[0], w[1], w[2])
        if y_pred != record[2]:
            error.append((record, y_pred))
    error_percent = int((len(error)/len(data))*100)
    epoch_accuracy = 100-error_percent
    print(f'Accuracy(%): {epoch_accuracy}\nError(%): {error_percent}\nUpdated Weights: {w[0]}, {w[1]}, {w[2]}\nLearning Rate: {lr}')
    print('\n')   
```
